We first need to create a folium `Map` object, with an initial center location to be NASA Johnson Space Center at Houston, Texas.

Now, let's add a circle for each launch site in data frame `launch_sites`


*TODO:*  Create and add `folium.Circle` and `folium.Marker` for each launch site on the site map


An example of folium.Circle:


`folium.Circle(coordinate, radius=1000, color='#000000', fill=True).add_child(folium.Popup(...))`


An example of folium.Marker:


`folium.map.Marker(coordinate, icon=DivIcon(icon_size=(20,20),icon_anchor=(0,0), html='<div style="font-size: 12; color:#d35400;"><b>%s</b></div>' % 'label', ))`


```python
launch_sites_df = spacex_df.groupby(['Launch Site'], as_index=False).first()[['Launch Site', 'Lat', 'Long']]

def create_launch_sites_map(launch_sites_df):
    # Center map around average coordinates
    map_center = [launch_sites_df['Lat'].mean(), launch_sites_df['Long'].mean()]
    site_map = folium.Map(location=map_center, zoom_start=4)
    
    # Add circle and marker for each launch site
    for idx, row in launch_sites_df.iterrows():
        coord = (row['Lat'], row['Long'])
        folium.Circle(
            location=coord,
            radius=1000,
            color='#d35400',
            fill=True
        ).add_child(folium.Popup(row['Launch Site'])).add_to(site_map)
        folium.map.Marker(
            location=coord,
            icon=DivIcon(
                icon_size=(20, 20),
                icon_anchor=(0, 0),
                html=f'<div style="font-size: 12; color:#d35400;"><b>{row["Launch Site"]}</b></div>'
            )
        ).add_to(site_map)
    return site_map

# Create the map
map_object = create_launch_sites_map(launch_sites_df)

# Display the map inline in Jupyter
display(map_object)
```

```python
# Task 2: Mark the success/failed launches for each site on the map

def create_launch_success_map(spacex_df):
    map_center = [spacex_df['Lat'].mean(), spacex_df['Long'].mean()]
    launch_map = folium.Map(location=map_center, zoom_start=4)
    
    marker_cluster = MarkerCluster().add_to(launch_map)
    
    for idx, row in spacex_df.iterrows():
        coord = (row['Lat'], row['Long'])
        color = 'green' if row['class'] == 1 else 'red'
        folium.CircleMarker(
            location=coord,
            radius=8,
            color=color,
            fill=True,
            fill_color=color,
            fill_opacity=0.9,
            popup=f"{row['Launch Site']} - {'Success' if row['class'] == 1 else 'Failure'}"
        ).add_to(marker_cluster)
    
    return launch_map

# Create the map
launch_map = create_launch_success_map(spacex_df)

# Display the map inline in Jupyter Notebook
display(launch_map)
# Example usage after loading spacex_df including 'class', 'Lat', 'Long', 'Launch Site'
# launch_map = create_launch_success_map(spacex_df)
# display(launch_map)
```

Next, we need to explore and analyze the proximities of launch sites.


Let's first add a `MousePosition` on the map to get coordinate for a mouse over a point on the map. As such, while you are exploring the map, you can easily find the coordinates of any points of interests (such as railway)


```python
# Initialize map centered on launch sites average location
map_center = [spacex_df['Lat'].mean(), spacex_df['Long'].mean()]
site_map = folium.Map(location=map_center, zoom_start=4)

# Create and add MarkerCluster
marker_cluster = MarkerCluster()
for idx, row in spacex_df.iterrows():
    coord = (row['Lat'], row['Long'])
    popup_text = f"{row['Launch Site']} - {'Success' if row['class'] == 1 else 'Failure'}"
    folium.Marker(
        location=coord,
        popup=popup_text,
        icon=folium.Icon(color=row['marker_color'])
    ).add_to(marker_cluster)
marker_cluster.add_to(site_map)

# Add MousePosition plugin to same map
formatter = "function(num) {return L.Util.formatNum(num, 5);};"
mouse_position = MousePosition(
    position='topright',
    separator=' Long: ',
    empty_string='NaN',
    lng_first=False,
    num_digits=5,
    prefix='Lat:',
    lat_formatter=formatter,
    lng_formatter=formatter,
)
site_map.add_child(mouse_position)

# Display combined map
display(site_map)
```

```python
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
```

*TODO:* Mark down a point on the closest coastline using MousePosition and calculate the distance between the coastline point and the launch site.


```python
# find coordinate of the closet coastline
# e.g.,: Lat: 28.56367  Lon: -80.57163
# distance_coastline = calculate_distance(launch_site_lat, launch_site_lon, coastline_lat, coastline_lon)
```

```python
# Launch site coordinate (e.g., Kennedy Space Center)
launch_site_coord = (28.5623, -80.5774)

# Coastline coordinate (provided)
coastline_coord = (28.56353, -80.568)

# Calculate distance between launch site and coastline
distance_coastline = calculate_distance(
    launch_site_coord[0], launch_site_coord[1],
    coastline_coord[0], coastline_coord[1]
)

# Center map between launch site and coastline
mid_lat = (launch_site_coord[0] + coastline_coord[0]) / 2
mid_lon = (launch_site_coord[1] + coastline_coord[1]) / 2
site_map = folium.Map(location=[mid_lat, mid_lon], zoom_start=15)

# Marker for launch site
folium.Marker(
    location=launch_site_coord,
    popup='Launch Site',
    icon=folium.Icon(color='blue', icon='rocket', prefix='fa')
).add_to(site_map)

# Marker for coastline point
folium.Marker(
    location=coastline_coord,
    popup='Closest Coastline',
    icon=folium.Icon(color='green', icon='info-sign')
).add_to(site_map)

# Marker with distance label near coastline point
distance_marker = folium.Marker(
    location=coastline_coord,
    icon=DivIcon(
        icon_size=(120, 30),
        icon_anchor=(0, 0),
        html='<div style="font-size: 14px; color:#d35400; background-color: white; border-radius: 3px; padding: 2px 4px;">'
             f'<b>{distance_coastline:.2f} KM</b></div>'
    )
)
distance_marker.add_to(site_map)

# Draw line connecting launch site and coastline point
line = folium.PolyLine(locations=[launch_site_coord, coastline_coord], weight=2, color='#d35400')
line.add_to(site_map)

# Display the map
display(site_map)

```
